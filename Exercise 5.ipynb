```python
release_dates = pd.read_csv('release_dates.csv', parse_dates=['date_published'], infer_datetime_format=True)
release_dates.head()
```

```python
# Make a bar plot of the months in which movies with "Christmas" in their title tend to be released in the USA.

rd = release_dates
rd = rd[rd.title.str.contains('Christmas')]
rd = rd[rd.country == 'USA']
rd.date_published.dt.month.value_counts().sort_index().plot(kind='bar')
```
